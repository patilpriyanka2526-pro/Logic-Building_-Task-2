```python
# input
sentence = input("Enter sentence: ")

# Convert sentence into list of words
words = sentence.lower().split()

# Convert list to set to remove duplicates
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Store employee names and salaries in dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

# Print result
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

```python
#  input in list format
numbers = eval(input("numbers: "))

# Initialize max and min
maximum = numbers[0]
minimum = numbers[0]

# Loop to find max and min
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Output
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
prices = input("prices: ")
prices = prices.strip("[]")
prices = list(map(int, prices.split(",")))

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
def calculate_attendance_percentage(attendance):
    total_days = len(attendance)  # Total number of days
    if total_days == 0:
        return 0.0  # Avoid division by zero

    present_count = 0  # Counter for present days
    for status in attendance:
        if status.upper() == "P":  # Check if present
            present_count += 1

    return (present_count / total_days) * 100  # Calculate percentage


def main():
    user_input = input('attendance: ')  # Take input
    user_input = user_input.strip()[1:-1]  # Remove brackets
    attendance = [item.strip().strip('"').strip("'") for item in user_input.split(",")]  # Clean items

    percentage = calculate_attendance_percentage(attendance)

    print("Attendance Percentage:", percentage)


if __name__ == "__main__":
    main()
```

```python
def remove_duplicates(phone_numbers):
    return set(phone_numbers)  # Convert list to set to remove duplicates


def main():
    user_input = input("phone numbers: ")  # Take input
    user_input = user_input.strip()[1:-1]  # Remove brackets
    phone_numbers = [int(num.strip()) for num in user_input.split(",")]  # Convert to list of integers

    unique_numbers = remove_duplicates(phone_numbers)  # Remove duplicates

    print("Unique phone numbers:", unique_numbers)  # Display result


if __name__ == "__main__":
    main()
```

```python
def count_character_frequency(text):
    frequency = {}  # Create empty dictionary

    for char in text:
        if char in frequency:
            frequency[char] += 1  # Increment count if character exists
        else:
            frequency[char] = 1  # Initialize count if new character

    return frequency


def main():
    user_input = input("Enter a string: ")  # Take string input from user
    result = count_character_frequency(user_input)  # Call function

    print(result)  # Display dictionary


if __name__ == "__main__":
    main()
```

```python
def convert_list_to_tuple(data_list):
    return tuple(data_list)  # Convert list to tuple


def main():
    user_input = input("Enter numbers separated by space: ")  # Take input
    data_list = [int(num) for num in user_input.split()]  # Convert input to list

    result = convert_list_to_tuple(data_list)  # Convert to tuple

    print("Tuple:", result)  # Display result


if __name__ == "__main__":
    main()
```

```python
def check_employee_exists(data_dict, name):
    employees = [emp.lower() for emp in data_dict["Employee"]]  # Convert all names to lowercase
    if name.lower() in employees:  # Compare in lowercase
        return "Employee exists"
    else:
        return "Employee does not exist"


def main():
    data = {
        "Employee": ["John", "Smith", "David"]  # Employee list
    }

    search_name = input("Enter name to search: ")  # User input

    result = check_employee_exists(data, search_name)
    print(result)


if __name__ == "__main__":
    main()
```

```python
def calculate_average(marks):
    total = sum(marks)  # Calculate sum of marks
    count = len(marks)  # Total number of subjects
    
    if count == 0:
        return 0  # Avoid division by zero
    
    return total / count  # Calculate average


def main():
    user_input = input("Enter marks separated by space: ")  # Take input
    marks = [float(mark) for mark in user_input.split()]  # Convert to list

    average = calculate_average(marks)  # Call function

    print("Average Marks:", average)  # Display result


if __name__ == "__main__":
    main()
```
